```python
import datetime
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import matplotlib.cbook as cbook

fig, ax = plt.subplots()
plt.title('Performance over the period')
ax2 = ax.twinx()

dfnew = w_df1.reset_index().set_index('Dates')
# dfnew.Rev_Pol.plot(color='red', ax=ax2)
dfnew.Rev_Pol.plot.bar(color='None', edgecolor='red', hatch='/', ax=ax2)
dfnew.Rating.plot(color='blue', ax=ax)


ax.set_ylabel(ylabel='Ratings', color='blue')
ax.set_xlabel('Period')
ax2.set_ylabel("Polarity", color='red')
ax.legend(bbox_to_anchor=(1.1,1), loc="upper left")
ax2.legend(bbox_to_anchor=(1.1,0.9), loc="upper left")
ax.set_xticklabels(labels = dfnew.index, rotation=75)
plt.show()
```
